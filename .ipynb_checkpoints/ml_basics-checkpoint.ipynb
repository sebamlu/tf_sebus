# LR


```python
import numpy as np 
import matplotlib.pyplot as plt

x_train = np.linspace(-1,1, 101)
y_train = 2 * x_train + np.random.randn(*x_train.shape) * 0.33

plt.scatter(x_train,y_train)
plt.show()
```

```python
import tensorflow as tf

lr = 0.01
epochs = 100
# model

x_input = tf.placeholder(dtype=tf.float32)
y_input = tf.placeholder(dtype=tf.float32)
w = tf.Variable(0.0, name='weights')
# LR model
y_hat = tf.multiply(x_input, w)

loss = tf.square(y_input - y_hat)
train_op = tf.train.GradientDescentOptimizer(lr).minimize(loss)
# Session stuff
sess = tf.Session()
init_op = tf.global_variables_initializer()
sess.run(init_op)

for epoch in range(epochs):
    for (x, y) in zip(x_train, y_train):
        sess.run(train_op, feed_dict={x_input: x, y_input: y})

w_star = sess.run(w)
plt.scatter(x_train, y_train)
plt.plot(x_train, x_train*w_star, 'r')
plt.show()
```
